```python
import pandas as pd
df_sampah = pd.read_excel('dataset_sampah.xlsx', sheet_name='data')

(df_sampah)
```

```python
total_2015=[]
for index, row in df_sampah.iterrows():
    if row['tahun']==2015:
        total_2015.append(row['jumlah_produksi_sampah'])

total_sampah_2015=sum(total_2015)
print(f"jumlah total sampah di tahun 2015: {total_sampah_2015}")
```

```python
hasil_kota_per_tahun = {}
for index, row in df_sampah.iterrows():
    tahun = row["tahun"]
    kota = row["nama_kabupaten_kota"]
    jumlah = row["jumlah_produksi_sampah"]

    if (tahun, kota) not in hasil_kota_per_tahun:
        hasil_kota_per_tahun[(tahun, kota)] = 0
    hasil_kota_per_tahun[(tahun, kota)] += jumlah

print("Produksi sampah per Kota/Kabupaten per tahun:")
for key, value in hasil_kota_per_tahun.items():
    print(f"{key[1]} ({key[0]}): {value} ton")
```

```python
from collections import defaultdict

hasil_kota_per_tahun = defaultdict(int)

# Iterasi langsung pada DataFrame
for index, row in df_sampah.iterrows():
    tahun, kota, jumlah = row["tahun"], row["nama_kabupaten_kota"], row["jumlah_produksi_sampah"]
    hasil_kota_per_tahun[(tahun, kota)] += jumlah

# Tampilkan hasil
print("Produksi sampah per Kota/Kabupaten per tahun:")
for (tahun, kota), jumlah in hasil_kota_per_tahun.items():
    print(f"{kota} ({tahun}): {jumlah} ton")
```
